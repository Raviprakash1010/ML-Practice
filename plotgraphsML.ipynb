```python
x=np.arange(0,20,2)
y=x**2
y2=x**3
```

```python
#plt.plot(x,y,color="red",marker="+")
plt.plot(x,y,color="red",marker="+")
```

```python
plt.ylabel("x^3")
plt.xlabel("x")
#plt.legend()
plt.title("Matplotlib demo")
plt.grid()
plt.text(2,80,"Text",fontsize=15)
plt.legend()
plt.plot(x,y,color="red",marker="+")
plt.plot(x,y2,color="black",marker="+")
plt.axis([0,10,0,10000])
plt.show()
```

```python
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population=[100,120,180,250,300,370]
```

```python
plt.scatter(year,salary,s=population,c="red",alpha=.5)
plt.title("Bubble chart")
plt.ylabel("salary")
plt.xlabel("year")
plt.legend()
plt.show()
```

```python
#ecah of point is different colour
c=np.arange(len(year))
plt.scatter(year,salary,s=population,c=c,alpha=.5,edgecolor="black",marker='^')
plt.title("Bubble chart")
plt.ylabel("salary")
plt.xlabel("year")
plt.legend()
plt.show()
```
